# Определение свойств легендарного покемона

Узнаем, связан ли первичный/вторичный тип покемона с его вероятностью оказаться легендарным 

```python
sd = df[['type1', 'type2', 'is_legendary']]
md = pd.get_dummies(sd)

corr = md.corr()
f, ax = plt.subplots(figsize=(13, 12))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Как мы можем видеть, тип не влияет на вероятность того, что покемон окажется легендарным

Давайте попробуем поработать с колонкой <b>capture_rate</b>

```python
df.capture_rate.unique()
```

Из выведенных данных видно, что каждый элемент столбца - это переменные типа string, но также есть элемент <i>'30 (Meteorite)255 (Core)'</i>. Давайте заменим его на 1000, чтобы в будущем не возникало проблем при обработке данных, а также заменим типы даннх с string на int.

```python
df['capture_rate'].replace('30 (Meteorite)255 (Core)', '1000', inplace=True)
pd.to_numeric(df['capture_rate'])
df['capture_rate'] = df['capture_rate'].astype(int)
df.capture_rate.unique()
```

```python
corr = df[['percentage_male', 'capture_rate','defense','sp_defense','base_happiness', 'speed', 'hp','attack','sp_attack','base_egg_steps','experience_growth','is_legendary']].corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

Исходя из полученной матрицы:
 <br><b>base_egg_steps</b> имеет высокую корелляцию с вероятностью легендарности покемона
 <br><b>base_happiness, percentage_male</b> и <b>capture_rate</b> имеют негативную корелляцию
 
 Взглянем на отобранные параметры поближе:

```python
corr = df[['percentage_male', 'capture_rate','base_happiness','base_egg_steps','is_legendary']].corr()
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
           annot=True)
```

Похоже, что именно <b>base_egg_steps</b> является ключевым критерием для определения легендарности покемона. 

Посмотрим на базовое количество шагов у всех легендарных покемонов:

```python
ax = sns.FacetGrid(df, col='is_legendary')
ax = ax.map(plt.hist, 'base_egg_steps')
```
